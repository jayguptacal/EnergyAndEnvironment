# Project Title: Predict Rock Density based on X-Ray Signal Strength using a Variety of Different Machine Learning Algorithms

## Project Background

The main goal is to predict rock density based on the X-ray signal strength using a variety of machine learning algorithms so that tunnel boring machine cutter head could be decided and planned proactively. While building this model, we can use a variety of machine learning algorithms and decide which one would be the best option.

Here is the summary of machine learning algorithms which will be used on this project:

- Linear Regression
- Polynomial Regression
- KNN Regressor
- Decision Tree Regressor
- Support Vector Regressor
- Random Forest Regressor
- Gradient Boosting Regressor
- AdaBoost Regressor

## Project Methodology

# Problem Identification

**Problem Hypothesis:** Predict rock density based on the X-ray signal strength using a variety of machine learning algorithms.

**Context for this Problem:** Since changing the tunnel boring machine cutter to suit the rock type can be an expensive challenge at the last minute, it is better to predict it and then plan for it. With so many machine learning options, the goal is to check a variety of different machine learning algorithms and see which one would be the best option.

# Data Wrangling

## Identification and Import of Data Set

The dataset is made up of X-ray signal vs. rock density to decide what type of tunnel boring cutting shield would be the best during tunnel boring process. Imagine if we are boring a new tunnel and we want to predict the best tunnel boring cutting shield based on different type of rocks so that we can minimize the downtime by proactively planning for changing boring cutting shield based on the machine learning modeling.

The dataset is available on my desktop and I will use it for the modeling purpose.

## Loading of Core Packages

```python
# Loading of core packages
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore')
```

```python
# Loading and reading the dataset
df = pd.read_csv('rock_density_xray.csv')

df.head()  
```

```python
# Column names are too long, so changing it here
df.columns = ['signal','density']
```

```python
# Checking for data type
df.info()
```

# Exploratory Data Analysis (EDA)

```python
# Checking for data with describe()
df.describe().T
```

```python
# Checking for the missing values
df.isnull().sum()
```

```python
# Plotting a scatter plot
sns.scatterplot(x='signal', y='density', data=df)
```

This graph shows that the relationship between signal and density is not linear and if we try to run a line, it will not be a good solution as the relationship is of sin-curve type.

Here density is the label and signal is one feature which will help predict the rock density. Also no scaling is required due to simple signal feature which is from 0 to 100.

```python
# Checking for the pairplot
sns.pairplot(df);
```

Because we have only one feature-driven dataset, we can move now to pre-processing stage.

# Pre-Processing and Training Data Development

```python
# Importing required packages
from sklearn.model_selection import train_test_split
```

```python
# Defining X and reshaping it as it is only 1 feature here
X = df['signal'].values.reshape(-1,1)
y = df['density']
```

```python
# splitting between training (90%) and testing set (10%)
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1, random_state=101)
```

# Machine Learning Modeling

Since the problem is quite unique, I am going to try a variety of different machine learning algorithm to check which one works the best for this scenario. Here is the list of the machine learning algorithms I will be testing:

- Linear Regression
- Polynomial Regression
- KNN Regressor
- Decision Tree Regressor
- Support Vector Regressor
- Random Forest Regressor
- Gradient Boosting Regressor
- AdaBoost Regressor

## Linear Regression

```python
# Importing required packages
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Creating linear regression model
lr_model = LinearRegression()

# fitting into the model
lr_model.fit(X_train,y_train)
```

```python
# predicting values
lr_preds = lr_model.predict(X_test)
```

```python
# mean absolute error for linear regression
mean_absolute_error(y_test,lr_preds)
```

```python
# mean squared error for linear regression
np.sqrt(mean_squared_error(y_test,lr_preds))
```

From the mean absolute error and mean squared error, this looks quite normal but let's check it out further if linear model is predicting the right values or not.

```python
# printing lr_preds values
lr_preds
```

There is something wrong here as all the values are around 2.2 value. So I need to dive into this further to see what's going on with the linear regression modeling. Since signal values are from 0 to 100, I can get predicted values for these values and see what are the predicted values.

```python
# creating a range for the x values
signal_range = np.arange(0,100)

# generating predicted values for these x-values using the linear regression model

density_preds = lr_model.predict(signal_range.reshape(-1,1))
```

```python
# plotting original values
plt.figure(figsize=(3,3),dpi=200)
sns.scatterplot(x='signal',y='density', data=df, color='red')

# plotting newly predicted data on the same above plot
plt.plot(signal_range,density_preds)
plt.show()
```

This shows that no matter what value we provide to signal, it is predicting a constant line values for density. Hence linear regression model is not working for this scenario. I am going to check for the polynomial regression modeling to see if that works for this scenario or not.

Before building the polynomial regression modeling, I am going to define a function to build different models, fit the model to training data, predict values using the model and then plotting values. This is an important step as this will help in building different algorithms faster.

```python
# defining the function
def run_model(model,X_train,y_train,X_test,y_test):
    # Fit model to training data
    model.fit(X_train,y_train)
    # predict values using the above model
    preds = model.predict(X_test)
    
    # Using root mean squared error for evaluation
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    print(f'RMSE: {rmse}')
    
    # Using mean absolute error for evaluation
    mae = mean_absolute_error(y_test,preds)
    print(f'MAE: {mae}')
    
    # Plot results for signal range 0-100
    signal_range = np.arange(0,100)
    output = model.predict(signal_range.reshape(-1,1))
    
    # plotting original values
    plt.figure(figsize=(3,3),dpi=200)
    sns.scatterplot(x='signal',y='density', data=df, color='red')

    # plotting newly predicted data on the same above plot
    plt.plot(signal_range,output)
    plt.show()    
```

## Polynomial Regression with Pipeline (degree=2)

```python
# Importing required packages
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
# creating pipeline for polynomial model
pipe_poly2 = make_pipeline(PolynomialFeatures(degree=2),LinearRegression())
```

```python
# Using helper function to run for the polynomial regression
run_model(pipe_poly2,X_train,y_train,X_test,y_test)
```

## Polynomial Regression with Pipeline (degree=6)

```python
# I am going to experiment now with a higher degree=6
pipe_poly6 = make_pipeline(PolynomialFeatures(degree=6),LinearRegression())

# Using helper function to run for the polynomial regression
run_model(pipe_poly6,X_train,y_train,X_test,y_test)
```

From the above model it looks that polynomial regression with degree=6 is a good fit for signal from 0 to 100 but we don't know if that is sufficient or it needs to be done for a higher range. Hence I am I am going to experiment with other algorithms now.

## K-Nearest Neighbors (KNN) Regressor Modeling

```python
# Importing required packages
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Testing different k-values here
k_values = [1,5,10,20]

for n in k_values:
    model = KNeighborsRegressor(n_neighbors=n)
    run_model(model,X_train,y_train,X_test,y_test)
```

```python
# Checking for a higher k-value like k=50
model_k = KNeighborsRegressor(n_neighbors=50)
run_model(model_k,X_train,y_train,X_test,y_test)
```

Higher the k-value, smoother the curve is for the KNN Regressor. But it is still not a good choice.

## Decision Tree Regressor Modeling

```python
# Importing required packages
from sklearn.tree import DecisionTreeRegressor
```

```python
# Running decision tree regressor with default hyperparameters as there is only 1 feature in the dataset
model = DecisionTreeRegressor()
run_model(model,X_train,y_train,X_test,y_test)
```

Like the lower k-values in the KNN Regressor, here the model is picking a lot of noise and trying to fit different values. So this is not a good choice.

## Support Vector Machine (SVM) Regressor Modeling with GridSearchCV

```python
# Importing required packages including GridSearchCV as it is important to do it for SVM
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
# Building the grid for SVM
svr = SVR()
param_grid = {
    'C':[0.01,0.1,1,5,10,100,1000],
    'gamma':['auto','scale']}

grid_svm = GridSearchCV(svr,param_grid)
```

```python
# checking performance for svm model
run_model(grid_svm,X_train,y_train,X_test,y_test)
```

Overall this is a good performance with the Support Vector Machine Regressor modeling.

## Random Forest Regressor Modeling

```python
# Importing required packages
from sklearn.ensemble import RandomForestRegressor
```

```python
# model for the random forest regressor
rfr = RandomForestRegressor(n_estimators=100)

run_model(rfr,X_train,y_train,X_test,y_test)
```

It is trying to overfit the values. So not a good performance with the random forest regressor modeling.

## Gradient Boosting Regressor Modeling

```python
# Importing required packages
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Running model for Gradient Boosting Regressor
model_gb = GradientBoostingRegressor()

run_model(model_gb,X_train,y_train,X_test,y_test)
```

Gradient Boosting model is also picking a lot of noise so this is not a good option.

## AdaBoost Regression Modeling

```python
# Importing required packages
from sklearn.ensemble import AdaBoostRegressor
```

```python
# Running model for AdBoost Regressor
model_ab = AdaBoostRegressor()

run_model(model_ab,X_train,y_train,X_test,y_test)
```

Based on the above, adaboost is also not a good option.
