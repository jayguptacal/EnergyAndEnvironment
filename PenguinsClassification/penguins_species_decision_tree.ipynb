# Project Title: Using Decision Tree Classifier, Predict Penguin Species based on the Penguin's Body Measurements 

## Project Background

Scientists have been collecting data on the Palmer archipelago, Antarctica for the different type of penguins. You can read more about this data here as data were collected and made available by Dr. Kristen Gorman (https://www.uaf.edu/cfos/people/faculty/detail/kristen-gorman.php) and the Palmer Station, Antarctica LTER (https://pal.lternet.edu/), a member of the Long Term Ecological Research Network (https://lternet.edu/).

Thank you to Dr. Gorman, Palmer Station LTER and the LTER Network! Special thanks to Marty Downs (Director, LTER Network Office) for help regarding the data license & use.

**The goal of this project is to use the Decision Tree, a supervised learning algorithm to identify the correct penguin species based on the penguin features.**

## Project Methodology

# Problem Identification

**Hypothesis:** Identify and predict a penguin species utilizing Decision Tree, a supervised machine learning algorithm based on the prenguin features.

**Context for Penguins Species Identification:** As part of their scientific studies, scientists in Antartica are identifying penguin species and monitoring their population. By taking their measurements of features such as weight, body length, beak length etc, scientists are tyring to predict correct penguin species based on using a machine learning algorithm.

# Data Wrangling

## Identification and Import of Dataset

Data for the penguins species is available as open source on various location including github (https://github.com/allisonhorst/palmerpenguins/blob/master/README.md) and kaggle (https://www.kaggle.com/parulpandey/palmer-archipelago-antarctica-penguin-data). I have downloaded it and it is now available on my desktop in the csv format.

## Data Description

The data contains following variables:

- species: penguin species (Chinstrap, Adlie, or Gentoo)
- culmen_length_mm: culmen length (mm)
- culmen_depth_mm: culmen depth (mm)
- flipper_length_mm: flipper length (mm)
- body_mass_g: body mass (g)
- island: island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
- sex: penguin sex

Here culmen is "the upper ridge of a bird's beak" (definition from Oxford Languages). For this penguin data, the culmen length and culmen depth are measured as shown below:



## Loading of Packages and Libraries

```python
# Importing required core packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading and reading the dataset
df = pd.read_csv('penguins_size.csv')

df.head()
```

```python
# Checking for data type
df.info()
```

```python
# Checking for unique species
df['species'].unique()
```

Based on the above, there are 3 Penguin species i.e. 'Adelie', 'Chinstrap', and 'Gentoo'. This is the target feature for this project and this problem belongs to multi-classification supervised machine learning algorithm.

```python
df.isnull().sum()
```

Since there are a total of 344 rows in the dataset and missing values are not signficant, I am going to drop these rows.

It seems there is a missing sex informtaion and currently identified by a dot. I am going to find out how many are there in the data and see how we can replace this information with the right sex type.

```python
# Checking for '.' type in the sex column
df[df['sex']=='.']
```

There is only 1 row with the dot in the sex feature. We can try to find and replace the sex based on other features provided here.

```python
# Checking for the pattern for male and female using describe() function for 'Gentoo' species only
df[df['species']=='Gentoo'].groupby('sex').describe()
```

Since the weight of the missing 'sex' penguin feature is closer to average female penguins, and even all other features are closer to female so it is safe to assign it as 'female'.

```python
# Using at method to change this dot value for the index 336
df.at[336,'sex']='FEMALE'
```

```python
# Checking if this value is change
df.loc[336]
```

```python
# Visualizing the data
sns.pairplot(df,hue='species')
```

```python
# Since separation seems very nice among the species with some overlap for features, I am going to run catplot now
sns.catplot(x='species',y='culmen_length_mm',data=df,kind='box', col='sex');
```

# Pre-Processing and Training Data Development

Since this is a **multi-label classification** problem, we need to use get_dummies() to change the categorical values for the decision tree analysis. I can do this for all the features except for the target feature of species one.

```python
# Using get_dummies for categorical features except for the target feature and categorizing this as X
X = pd.get_dummies(df.drop('species', axis=1),drop_first=True)
```

```python
# Defining y here
y = df['species']
```

**Important Reminder:** Since I am using decision tree algorithm, I don't need to use any type of features scaling as decision tree uses one feature at a time for classification, hence no scaling is required.

```python
# Importing packages for training and testing data split
from sklearn.model_selection import train_test_split
```

```python
# Splitting X and y to create 30% for testing purpose
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=101)
```

# Machine Learning Modeling

## Decision Tree Classifier with Default Parameters

```python
# Importing required packages here
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
# Default hyper parameter
# help(DecisionTreeClassifier())
# Default parameters: DecisionTreeClassifier(*, criterion='gini', splitter='best', max_depth=None, 
# min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, 
# random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, min_impurity_split=None, 
# class_weight=None, ccp_alpha=0.0)
```

```python
# creating base model here
model = DecisionTreeClassifier()
```

```python
# fitting training data
model.fit(X_train, y_train)
base_preds = model.predict(X_test)
```

```python
# classification report
print("Classification report for base predictions:\n", classification_report(y_test, base_preds))
```

```python
# plotting confusion matrix
plot_confusion_matrix(model, X_test, y_test);
```

Using **base decision tree classifier with default hyperparameters**, there are only 2 penguins being miss-classified here.

```python
# Printing Feature importance matrix as dataframe
pd.DataFrame(index=X.columns,data=model.feature_importances_, columns=["Feature Importances"]).sort_values("Feature Importances", ascending=False)
```

```python
# visualize tree here with sklearn plot_tree
from sklearn.tree import plot_tree
```

```python
# Plotting tree
plt.figure(figsize=(12,8), dpi=200)
plot_tree(model, feature_names=X.columns, filled=True);
```

Before I start experimenting with the hyperparameters, I am going to setup a function to print a Classification report as well as plot the tree with one defined function.

```python
# defining the function
def report_model(model):
    model_preds = model.predict(X_test)
    # printing classifcation report
    print(classification_report(y_test, model_preds))
    print('\n')
    # plotting decision tree
    plt.figure(figsize=(12,8), dpi=200)
    plot_tree(model, feature_names=X.columns, filled=True);
```

## Decision Tree Classifier with Different Hyperparameters

### Decision Tree Classifier with max_depth=2

```python
# If we look at the base model, there is a max_depth of 5 and now experimenting with 2 only
pruned_tree = DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
```

```python
# reporting for the pruned tree
report_model(pruned_tree)
```

Even though the performance has decreased but now tree is much simpler and it is considering only 3 features now i.e. flipper_length_mm, culmen_length_mm and culmen_depth_mm.

### Decision Tree Classifier with max_leaf_nodes=3

```python
# creating a model with max leaf nodes
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)

# fitting training data to this model
max_leaf_tree.fit(X_train,y_train)
```

```python
# printing results for the model
report_model(max_leaf_tree)
```

### Decision Tree Classifier with criterion='entropy'

```python
# Rather than default gini criterion, now changing it to entropy
entropy_tree = DecisionTreeClassifier(criterion='entropy')

# fitting to this model
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```

### Decision Tree Classifier with GridSearchCV to Find Best Hyperparameters

```python
# Importing required packages
from sklearn.model_selection import GridSearchCV
```

```python
# Parameters for grid search cv
param_dict = {
    "criterion":["gini","entropy"],
    "max_depth":range(1,10),
    "min_samples_split":range(1,10),
    "min_samples_leaf":range(1,5)
    }
```

```python
# Using gridsearch to find best parameters
decision_tree = DecisionTreeClassifier()
grid = GridSearchCV(decision_tree,param_grid=param_dict,cv=10,n_jobs=-1)
grid.fit(X_train,y_train);
```

```python
# identifying best parameters
grid.best_params_
```

```python
# finding the best score
grid.best_score_
```

### Decision Tree Classifier with Best Hyperparameters

```python
best_tree = DecisionTreeClassifier(criterion='gini',max_depth=6,min_samples_leaf=1,min_samples_split=2)
best_tree.fit(X_train,y_train)
```

```python
report_model(best_tree)
```

```python
# Plot confusion maatrix for the best parameters
plot_confusion_matrix(best_tree,X_test,y_test);
```
